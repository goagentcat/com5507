# STEP 2  Joining csv files
Each CSV has same structure but different "range" or "parts" from the previous data scraping process

```python
import requests 
import bs4 as bs 
import time, json, random
import pandas as pd 
import csv 
d1 = pd.read_csv("allFastFood-nutriData0,500.csv")
d2 = pd.read_csv("allFastFood-nutriData500,1000.csv")
d3 = pd.read_csv("allFastFood-nutriData1000,1500.csv")
d4 = pd.read_csv("allFastFood-nutriData1500,2000.csv")
d5 = pd.read_csv("allFastFood-nutriData2000,2500.csv")
d6 = pd.read_csv("allFastFood-nutriData2500,3000.csv")
d7 = pd.read_csv("allFastFood-nutriData3000,3500.csv")
d8 = pd.read_csv("allFastFood-nutriData3500,4000.csv")
d9 = pd.read_csv("allFastFood-nutriData4000,4153.csv")

frames=[d1,d2,d3,d4,d5,d6,d7,d8,d9]
res=pd.concat(frames)
res.reset_index(inplace=True)
res.drop(res.columns[0], axis=1, inplace=True)
res.drop(res.columns[0], axis=1, inplace=True) #drop column 0 another time to delete the extra index column
res.to_csv("allFastFood-nutriData-all.csv")
res
```
