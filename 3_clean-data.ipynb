```python
import requests 
import bs4 as bs 
import time, json, random
import pandas as pd 
import csv 
data = pd.read_csv("allFastFood-nutriData-all.csv")
 

#data.head(5)

    
# list(data_top) or 
dataIndexes=list(data.index) 

dropList=[]

fail=0
for i in dataIndexes:
    if data["Health_Rating"][i]==-9999 or data["Taste_Rating"][i]==-9999 or data["Total_Fat"][i]==-9999 or \
    data["Saturated_Fat"][i]==-9999 or data["Calories"][i]==-9999 or  data["Sugars"][i]==-9999 or\
    data["Dietary_Fiber"][i]==-9999 or data["Sodium"][i]==-9999 or  data["Total_Fat_PDV"][i]==-9999 or\
    data["Saturated_Fat_PDV"][i]==-9999 or data["Cholesterol_PDV"][i]==-9999 or  data["Sodium_PDV"][i]==-9999 or\
    data["Total_Carbohydrates_PDV"][i]==-9999 or data["Dietary_Fiber_PDV"][i]==-9999 or  data["Protein_PDV"][i]==-9999 :
        fail=fail+1
        print("Not complete data- "+ data["Brand"][i]+" "+data["Product"][i])
        dropList.append(i)


print(str(fail)+" out of "+str(len(dataIndexes))+" has incomplete data ")


data = data.drop(dropList, axis=0)
data.reset_index(inplace=True)
data.drop(data.columns[0], axis=1, inplace=True)
data.drop(data.columns[0], axis=1, inplace=True)

data

data.to_csv("allFastFood-clean.csv")
#return (df)
 
```
