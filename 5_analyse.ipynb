### Below are statistic of food in different restaurants

```python
#% of healthy food in one resautrant
Brand="KFC"
All=data[(data['Brand']==Brand)]
Part=data[(data['Brand']==Brand) & (data['USA_Female_Below_40%DV']=="Y") & (data['USA_Male_Below_40%DV']=='Y')&(data['China_Female_Below_40%DV']=='Y')\
& (data['China_Male_Below_40%DV']=="Y") & (data['Europe_Female_Below_40%DV']=='Y')&(data['Europe_Male_Below_40%DV']=='Y')]

print(Brand+" has "+str(round(len(Part)/len(All)*100, 2))+"% healthier food")
All[['Calories','Total_Fat','Sugars','Sodium','Dietary_Fiber']].describe()
```
